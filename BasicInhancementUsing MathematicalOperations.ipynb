```python
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from IPython.display import Image
# from PIL import Image
import cv2
import numpy as np
```

```python
z = cv2.imread(r"sample121.jpg", cv2.IMREAD_COLOR)
zz = cv2.cvtColor(z, cv2.COLOR_RGB2BGR)
Image('sample121.jpg')
```

```python
#           Adjusting Brightness
a = np.ones(z.shape, dtype="uint8") * 50
x = cv2.add(z,a)
zz = cv2.subtract(z,a)
plt.figure(figsize=[20,5])
plt.subplot(141); plt.title("brighter"); plt.imshow(x)
plt.subplot(142); plt.title("Original"); plt.imshow(z)
plt.subplot(143); plt.title("darker"); plt.imshow(zz)
```

```python
#        Contrast or Multiplication

Imagee = cv2.imread(r'sample121.jpg', cv2.IMREAD_COLOR)
imagee = cv2.cvtColor(Imagee, cv2.COLOR_RGB2BGR)

for_low = np.ones(imagee.shape, dtype='uint8') * .8
for_high = np.ones(imagee.shape, dtype = 'uint8') * 1.2

low = np.uint8(cv2.multiply(np.float64(imagee), for_low))
high = np.uint8(cv2.multiply(np.float64(imagee), for_high))

plt.figure(figsize=(30,20))
plt.subplot(141); plt.imshow(low); plt.title("Low Contrast")
plt.subplot(142); plt.imshow(imagee); plt.title('Original')
plt.subplot(143); plt.imshow(high); plt.title('High Contrast')
```

```python
# To put a limit to fix the contrast or to remove the 0 elements on matrix/Image or the red circle

high = np.uint8(cv2.multiply(np.float64(imagee), for_high))
circle = cv2.circle(high, (150,100), 200, (255, 54, 232), 10)
plt.figure(figsize = (30,10))
plt.subplot(141); plt.imshow(high); plt.title('Default High Contrast')
fixed_contrast = np.uint8(np.clip(cv2.multiply(np.float64(imagee), for_high), 0, 255))
circle2 = cv2.circle(fixed_contrast, (150,100), 200, (255, 54, 232), 10)
plt.subplot(142); plt.imshow(circle2); plt.title('Fixed Contrast')
```

```python
#       Thresholding - Used in order to produce a binary map
openn = cv2.imread(r'building-architecture-windows-city-royalty-free-thumbnail.jpg', cv2.IMREAD_GRAYSCALE)
retval, thresh = cv2.threshold(openn, 100, 255, cv2.THRESH_BINARY) # File, Threshold Amount, Limit, Type

plt.figure(figsize=(20,5))
plt.subplot(121); plt.imshow(openn, cmap="gray"); plt.title('Original')
plt.subplot(122); plt.imshow(thresh, cmap = 'gray') ; plt.title('Applied Threshold')
```

```python
piano_sheet = cv2.imread(r"png-transparent-sheet-music-piano-musical-note-song-sheet-music-thumbnail.png", cv2.IMREAD_GRAYSCALE)
grayed = cv2.cvtColor(piano_sheet, cv2.COLOR_RGB2BGR)
retval, piano_threshold = cv2.threshold(grayed, 50, 255, cv2.THRESH_BINARY)
retval, another1 = cv2.threshold(grayed, 130, 255, cv2.THRESH_BINARY)
applied = cv2.adaptiveThreshold(grayed, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 7)
plt.figure(figsize = (20,5))
plt.subclass(141); plt.imshow(grayed); plt.title('Original')
plt.subclass(142); plt.imshow(piano_threshold); plt.title("Threshold - 50")
plt.subclass(143); plt.imshow(another1); plt.title('Threshold - 130')
plt.subclass(144); plt.imshow(applied, cmap='gray'); plt.title('Thresholded (Adaptive)')
```
